```python
import matplotlib.pyplot as plot
import numpy as np


# Frequenzy on x-axis in kHz
#x = np.linspace(10**-2, 10**4, 20)
#import matplotlib.pyplot as plot
plot.style.use('classic')

# Base Voltage on Y-Axises
dotedkHz01 = [ 60]
dotedVolt01 = [0.48]

# Voltage on Y-Axises
dotedkHz1 = [ 60]
dotedVolt1 = [0.098]

# Voltage on Y-Axises
dotedVolt2 = [0.0002]

# Voltage on Y-Axises
dotedVolt3 = [0.00408]


dotedkHz = [10**-1.8, 10**1.3, 10**2, 10**3, 10**4]
dotedVolt = [10**0, 10**0, 10**0.7, 10**0.8, 10**4 ]


redkHz = [10**-2, 10**-2, 10**0, 10**4]
redVolt = [10**-1.1, 10**-1.5, 10**-1.55, 10**2]


kHz =     [10**-1.8, 1.1,  10**0.3, 10**2.4,  10**3,  10**4]
Voltage = [10**-2.7,    10**-1.7   ,10**-1.9,  1  ,   1,      10**-1]



# Display grid
ax = plot.subplot()
ax.tick_params(axis='both', which='major', labelsize=10)

params = {'legend.fontsize': 10,
          'legend.handlelength': 2}
plot.rcParams.update(params)

plot.grid(True, which="both")
# Linear X axis, Logarithmic Y axis
plot.semilogy(kHz,Voltage )
plot.semilogx(kHz,Voltage,'-g' , label = 'Stufe 1')
###############################
plot.semilogy(redkHz,redVolt )
plot.semilogx(redkHz,redVolt,'-r', label = 'Stufe 2' )
###################
plot.semilogy(dotedkHz,dotedVolt  )
plot.semilogx(dotedkHz,dotedVolt,'-k', label = 'Stufe 3' )

plot.semilogy(dotedkHz01,dotedVolt01  )
plot.semilogx(dotedkHz01,dotedVolt01,'ro', label = 'Base Voltage' )

plot.semilogy(dotedkHz1,dotedVolt1  )
plot.semilogx(dotedkHz1,dotedVolt1,marker='o', color='0.88', linestyle=':', label = 'Value 1' )


plot.semilogy(dotedkHz1,dotedVolt2  )
plot.semilogx(dotedkHz1,dotedVolt2,'go', label = 'Value 2' )

plot.semilogy(dotedkHz1,dotedVolt3  )
plot.semilogx(dotedkHz1,dotedVolt3,'bo', label = 'Value 3' )



plot.ylim([10**-3,21000])
plot.xlim([10**-2,10000])
# Provide the title for the semilog plot
plot.title('Simulation - Case 1')
# Give x axis label for the semilog plot
plot.xlabel('Frequency [kHz]')
# Give y axis label for the semilog plot
plot.ylabel('Spitze Wert kHZ[Volt] U$_{H}$')
plot.legend(framealpha=1, frameon=True, loc='upper right');
#ax.tick_params(axis='both', which='major', labelsize=10)


#plot.plot(85,.0041,'co')
#plot.plot(85,.00329,'ro')
#plot.plot(85,.28,'ro')
#plot.plot(85,.18,'ro')
#plot.plot(85,.08,'ro')
fig1 = plot.gcf()
plot.show()
fig1.savefig('WorstcaseUH', dpi=100)
```
